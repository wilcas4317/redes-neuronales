## Este colab es para la explicación de el entrenamiento de la primera red neuronal

Las variables w1, w2 representan los pesos de las neuronas, y b es el bias

```python
import pandas as pd

df = pd.read_csv("sample_data/data.csv")

frentes = df["frente"]
profundidades = df["profundidad"]
precios = df["precio"]

N = len(frentes)

def calculo_neurona(frente, profundidad, w1, w2, b):
  y = w1*frente + w2*profundidad + b
  return y

#calcula los errores de todos y luego los suma
def MSE(frentes, profundidades, precios, w1, w2, b):
  suma = 0

  for i in range(N):
    suma = suma + (  precios[i] - calculo_neurona(frentes[i], profundidades[i], w1, w2, b)  )**2

  return (1/(2*N))*suma

#derivada con respecto a w1
#(1/N) x suma (y-yb)*(-frente)
def derivadaW1(frentes, profundidades, precios, w1, w2, b):
  suma = 0

  for i in range(N):
    suma = suma + (precios[i] - calculo_neurona(frentes[i], profundidades[i], w1, w2, b))*(-frentes[i])

  return (1/N)*suma

#derivada con respecto a w1
#(1/N) x suma (y-yb)*(-profundidad)
def derivadaW2(frentes, profundidades, precios, w1, w2, b):
  suma = 0

  for i in range(N):
    suma = suma + (precios[i] - calculo_neurona(frentes[i], profundidades[i], w1, w2, b))*(-profundidades[i])

  return (1/N)*suma

#derivada con respecto a b
#(1/N) x suma (y-yb)*(-1)
def derivadaB(frentes, profundidades, precios, w1, w2, b):
  suma = 0

  for i in range(N):
    suma = suma + (precios[i] - calculo_neurona(frentes[i], profundidades[i], w1, w2, b))*(-1)

  return (1/N)*suma
```

# Ahora vamos a aplicar o modelar el descenso del gradiente

```python
#descenso del gradiente a través del proceso de entrenamiento

#fit
def entrenamiento():
  w1=0
  w2=0
  b=0

  n=0.0001

  epocas = 1000

  #Ahora si el descenso del gradiente
  for i in range(epocas):
    w1 = w1 - n*derivadaW1(frentes, profundidades, precios, w1, w2, b)
    w2 = w2 - n*derivadaW2(frentes, profundidades, precios, w1, w2, b)
    b = b - n*derivadaB(frentes, profundidades, precios, w1, w2, b)

    print("Error obtenido en la época ", i, MSE(frentes, profundidades, precios, w1, w2, b))

  print("---------------------------------")
  print("Mejores valores obtenidos: ")
  print("w1 ", w1)
  print("w2 ", w2)
  print("b ", b)
```

Ahora vamos a entrenar

```python
entrenamiento()
```
